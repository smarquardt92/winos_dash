```python
import pandas as pd
```

```python
df = pd.read_csv("us_wines_coords.csv")
df = df.drop(['country', 'latitude', 'longitude', 'Unnamed: 0'], axis=1)
df = df.rename(index=str, columns={"province": "state"})
df.head()
```

```python
price_df = df.drop(['points', 'variety'], axis=1)
price_df = price_df.rename(index=str, columns={"price": "value"})
price_df = price_df.groupby(['state']).mean()
price_df
```

```python
points_df = df.drop(['price', 'variety'], axis=1)
points_df = points_df.rename(index=str, columns={"points": "value"})
points_df = points_df.groupby(['state']).mean()
points_df
```

```python
variety_df = df.drop(['price'], axis=1)
variety_df = variety_df.groupby(['variety']).mean()
variety_df
```

```python
export_csv = points_df.to_csv("points_average.csv")
```
