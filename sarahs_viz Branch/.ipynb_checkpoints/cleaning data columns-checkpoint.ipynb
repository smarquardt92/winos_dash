## Cleaning 11 column dataset

```python
csv_file = "Resources/mag_columns_11.csv"
wine_df = pd.read_csv(csv_file)
wine_df.head()
```

```python
#create new data with select columns
new_wine11_df = wine_df[['description', 'points', 'country', 'province', 'variety', 'price' ]]
new_wine11_df.head()
```

```python
#clean data frame
new_wine11_df.count()
```

```python
#drop missing rows
new_wine11_df = new_wine11_df.dropna(how ="any")
```

```python
new_wine11_df.count()
```

```python
both = [new_wine_df, new_wine11_df]
combined_df = pd.concat(both)
```

```python
combined_df.head()
```

```python
combined_df.count()
```

```python
us_wine_df = combined_df[combined_df["country"] == "US"]
us_wine_df.head()
```

```python
us_wine_df.dropna(how="any")
```

```python
us_wine_df.count()
```

```python
us_wine_df.to_csv("Resources/us_wines.csv")
```

```python
#to_sql has the name, con if exists_ parameters, con is equal to the sqlalchemy engine, if exists tells pandas 
#what to do if the engine already exists
#we are passing sql alchemy into pandas
us_wine_df.to_sql(name='wine', con=engine, if_exists='append', index=False)
```
